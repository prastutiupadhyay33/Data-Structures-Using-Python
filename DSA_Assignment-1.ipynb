```python
import numpy as np

# Step 1: Initialize 3D array

data = np.arange(3*16*10).reshape(3,16,10)
print("Original Data Shape:",data.shape)
print(data)
```

```python
# Step 2: Year-wise average

year_avg = data.mean(axis=(1,2))
print("\nYear-wise Average:")
print(year_avg)
```

```python
# Step 3: Student-wise total

student_total=data.sum(axis=(0,2))
print("\nStudent-wise Total:")
print(student_total)
```

```python
# Step 4: Subject-wise highest score

subject_max=data.max(axis=(0,1))
print("\nSubject-wise Highest Score:")
print(subject_max)
```

```python
# Step 5: Replace marks below 40 with 40

data[data<40]=40
print("\nUpdated Data (Marks < 40 replaced with 40):")
print(data)
```

```python
# Step 6: Sort marks subject-wise

sorted_data = np.sort(data,axis=1)
print("\nSorted Data(Subject-wise):")
print(sorted_data)
```
