```python
import numpy as np
# 1. Calculator
def calculator(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b if b != 0 else "Division by zero error"
    elif op == '^':
        return a ** b
    else:
        return "Invalid operation"

a = float(input())
b = float(input())
op = input()

result = calculator(a, b, op)
print(result)
```

```python
def discount(price, discount_percent):
    return price - (price * discount_percent / 100)

price = float(input())
discount_percent = float(input())

final_price = discount(price, discount_percent)
print(final_price)
```

```python
def grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 75:
        return 'B'
    elif marks >= 60:
        return 'C'
    elif marks >= 40:
        return 'D'
    else:
        return 'F'

marks = float(input())

result = grade(marks)
print(result)
```

```python
def even_odd(num):
    return "Even" if num % 2 == 0 else "Odd"

num = int(input())

result = even_odd(num)
print(result)
```

```python
#Bubble sort
def BBS(array):
  n = len(array)
  for i in range(n-1):
    for j in range (n-1-i):
      if array[j] > array[j+1]:
        array[j], array[j+1] = array[j+1],array[j]
  return array

data= [7,6,5,4,3,2,1]
print(BBS(data))
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
              min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
        # print(arr)
    return arr
arr = [89,76,54,47,33,20,10,1]
print(selection_sort(arr))
```

```python
arr = [8,7,6,5,4,3,2,1]
for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key
print(arr)
```

```python
def quick_sort(arr):
    if len(arr) <= 1: 
        return arr
    pivot = arr[-1]   
    left = []
    right = []
    for i in range(len(arr) - 1):
        if arr[i] < pivot:
            left.append(arr[i])
        else:
            right.append(arr[i])
    return quick_sort(left) + [pivot] + quick_sort(right)

arr = [9,6,3,8,7,2]
sorted_arr = quick_sort(arr)
print("Sorted array:", sorted_arr)
```
